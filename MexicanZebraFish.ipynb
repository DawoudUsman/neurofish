```python
# Performing PCA -> Dimensionality reduction
sss = pd.concat([loom_pd, loom_right], axis=1).to_numpy()[:1000]
n, d = sss.shape
U, D = pca(sss)

#plot the square-root eigenvalue spectrum
fig = plt.figure()
plt.plot(D,'-o',color='black')
plt.xlabel('dimensions')
plt.ylabel('square-rooted eigenvalues')
plt.title('Square-root eigenvalue spectrum')
plt.show()
```

```python
percent_var = 100 * np.sum(D[:3])/np.sum(D)
print('variance explained = ', percent_var)
```

```python
colors = ['r','k','y','g','b','m','c','violet','darkviolet','purple']
_angle = 100;
fig = plt.figure()
ax  = fig.add_subplot(projection='3d')

for loomangle in range(n//_angle):
    indxx = np.arange(0,_angle,1)+(loomangle)*_angle
    X = sss[indxx,:]
    ax.scatter(X@U[:,0],X@U[:,1],X@U[:,2],'o', color=colors[loomangle])
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')

plt.title('Reach data projected into three-dimensional PC space')
```
